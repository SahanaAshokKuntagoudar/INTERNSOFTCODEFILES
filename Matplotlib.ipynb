## What is matplotlib?

Matplotlib is the most popular and mature library for plotting data using
Python. It has all of the functionality you would expect, including the ability to control
the formatting of plots and figures at a very fine level.

## Plotting in Pandas

```python
n=pd.Series(np.random.normal(size=10))
print(n)
n.plot()
```

```python
n.cumsum().plot(grid=True)
```

Similarly, for a DataFrame:

```python
var=pd.DataFrame({'normal':np.random.normal(size=10),
                 'gamma':np.random.gamma(1,size=10),
                 'poisson':np.random.poisson(size=10)})
print(var)
var.cumsum(0).plot()
```

```python
var.cumsum(0).plot(subplots=True)
```

```python
var.cumsum(0).plot(secondary_y='normal')
```

```python
fig,axes=plt.subplots(nrows=3, ncols=1, figsize=(4,12))
for i,v in enumerate(['normal','gamma','poisson']):
    var[v].cumsum(0).plot(ax=axes[i], title=v)
axes[0].set_ylabel('cumulative sum')
```

# Simple Line Plots

Perhaps the simplest of all plots is the visualization of a single function $y = f(x)$.

```python
fig=plt.figure()
ax=plt.axes()

a=np.linspace(0,40,100)
print(a)
ax.plot(a,np.sin(a));
```

```python
plt.plot(a,np.sin(a))
```

## Line Colors and Styles

```python
plt.plot(a,np.sin(a-0), color='red')
plt.plot(a,np.sin(a-1), color='y')
plt.plot(a,np.sin(a-2), color='0.55')
```

If no color is specified, Matplotlib will automatically cycle through a set of default colors for multiple lines.

```python
plt.plot(a,np.sin(a-0), linestyle='-')
plt.plot(a,np.sin(a-1), linestyle='-.')
plt.plot(a,np.sin(a-2), linestyle=':')
```

## Adjusting the Plot: Axes Limits

```python
plt.plot(a,np.sin(a-0))

plt.xlim(-1,12)
plt.ylim(-2,2)
```

## Labeling Plots

```python
plt.plot(a,np.cos(a))
plt.title("conscine curve")
plt.xlabel("x")
plt.ylabel("cos(x)")
```

## Scatter Plots with ``plt.plot``

```python
c=np.linspace(1,10,40)
d=np.log(c)

plt.plot(c,d,'o',color='green')
```

```python
plt.plot(c,d,'-or')
```

## Scatter Plots with ``plt.scatter``

A second, more powerful method of creating scatter plots is the ``plt.scatter`` function, which can be used very similarly to the ``plt.plot`` function:

```python
plt.scatter(c,d, marker='d')
```

```python
rng=np.random.RandomState(6)
x=rng.randn(300)
y=rng.randn(300)
colors=rng.rand(300)
size=1000*rng.rand(300)

plt.scatter(x,y,c=colors,s=size, alpha=0.5, cmap='Reds')
plt.colorbar()
```

# Visualizing Errors

## Errorbars

A basic errorbar can be created with a single Matplotlib function call:

```python
a=np.linspace(0,20,200)
dy=0.1
y=np.sin(a)+ dy *np.random.randn(200)

plt.errorbar(a,y,xerr=dy,fmt='.b');
```

```python
a=np.linspace(0,20,200)
dy=0.1
y=np.sin(a)+ dy *np.random.randn(200)

plt.errorbar(a,y,yerr=dy,fmt='.b');
```

## Two-Dimensional Histograms

```python
mean=[2,4]
cov=[[2,3],[9,8.0]]
x,y=np.random.multivariate_normal(mean,cov,10000).T
```

### ``plt.hist2d``: Two-dimensional histogram

```python
plt.hist2d(x,y,bins=10,cmap='Reds')
cb=plt.colorbar()
```

### ``plt.hexbin``: Hexagonal binnings

```python
plt.hexbin(x,y,gridsize=20,cmap='Reds')
cb=plt.colorbar()
```
